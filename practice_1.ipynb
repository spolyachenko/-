```r
library(lmtest)
data = swiss
help(swiss)
```

## Examination

```r
plot(data$Examination, ylab = "Examination")
print(paste("Mean: ", mean(data$Examination)))
print(paste("Variance: ", var(data$Examination)))
print(paste("Standard Deviation: ", sd(data$Examination)))
```

#### У набора Examination стандартное отклонение сравнимо со средним значением, значит значения Examination достаточно сильно распределены

## Agriculture

```r
plot(data$Agriculture, ylab = "Agriculture")
print(paste("Mean: ", mean(data$Agriculture)))
print(paste("Variance: ", var(data$Agriculture)))
print(paste("Standard Deviation: ", sd(data$Agriculture)))
```

#### У набора Agriculture стандартное отклонение сравнимо со средним значением, значит значения Agriculture достаточно сильно распределены

## Fertility

```r
plot(data$Fertility, ylab = "Fertility")
print(paste("Mean: ", mean(data$Fertility)))
print(paste("Variance: ", var(data$Fertility)))
print(paste("Standard Deviation: ", sd(data$Fertility)))
```

#### У набора Fertility стандартное отклонение ~в 5 раз меньше среднего значения, значит значения Fertility не сильно распределены

## Зависимость Examination от Agriculture

```r
model_Exam_from_Agricult = lm(Examination~Agriculture, data)
model_Exam_from_Agricult
plot(data$Agriculture, data$Examination, xlab = "Agriculture", ylab = "Examination")
summary(model_Exam_from_Agricult)
```

Зависимость: Examination = 28.7 - 0.24 * Agriculture

p-значение близко к нулю, три "звездочки", значит зависимость присутствует. 

Коэффициент детерминации равен 0.47, значит модель плохая

## Зависимость Examination от Fertility 

```r
model_Exam_from_Fert = lm(Examination~Fertility, data)
model_Exam_from_Fert
plot(data$Fertility, data$Examination, xlab = "Fertility", ylab = "Examination")
summary(model_Exam_from_Fert)
```

Зависимость: Examination = 45.4 - 0.41 * Fertlitiy
